```python
import matplotlib.pyplot as plt

men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]

ind = range(len(men_means))

plt.figure(figsize=(8, 6))
plt.bar(ind, men_means, yerr=men_std, label='Men', color='lightblue', alpha=0.7, capsize=5)
plt.bar(ind, women_means, bottom=men_means, yerr=women_std, label='Women', color='orange', alpha=0.7, capsize=5)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5'])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Data')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', facecolors='none', edgecolors='blue', alpha=0.7)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Empty Circles')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
radius = np.random.rand(50) * 100

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=radius, color='blue', alpha=0.7)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Balls of Different Sizes')

plt.grid(True)
plt.tight_layout()
plt.show()
```
