```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35, 40],
        'Gender': ['male', 'female', 'Male', 'FeMale']}

df = pd.DataFrame(data)

# Specified character column to swap cases
specified_column = 'Gender'

# Swap cases of the specified column
df[specified_column] = df[specified_column].str.swapcase()

print(df)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}
df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

start_date = '2016-10-03'
end_date = '2016-10-07'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.plot(filtered_df['Date'], filtered_df['Open'], label='Open')
plt.plot(filtered_df['Date'], filtered_df['High'], label='High')
plt.plot(filtered_df['Date'], filtered_df['Low'], label='Low')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Financial Data of Alphabet Inc. (Oct 3-7, 2016)')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y1_values = [2, 3, 5, 7, 11]
y2_values = [1, 4, 6, 8, 10]

# Plotting the lines
plt.plot(x_values, y1_values, label='Line 1', color='blue', linewidth=2)
plt.plot(x_values, y2_values, label='Line 2', color='red', linewidth=1.5)

# Adding legend
plt.legend()

# Adding labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2)

axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sin(x)')

axs[1].plot(x, y2, color='red')
axs[1].set_title('Cos(x)')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'JavaScript', 'C#', 'C++', 'PHP']
popularity = [30, 25, 20, 15, 10, 5]

plt.bar(languages, popularity, color='skyblue')

plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4']
scores_men = [20, 35, 30, 35]
scores_women = [25, 32, 34, 20]

# Define the width of the bars
bar_width = 0.35

# Define the X locations for the groups
x = np.arange(len(groups))

# Create the bar plot
plt.bar(x - bar_width/2, scores_men, width=bar_width, label='Men')
plt.bar(x + bar_width/2, scores_women, width=bar_width, label='Women')

# Adding labels and title
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(x, groups)  # Set group names as x-axis ticks
plt.legend()  # Add legend

# Displaying the plot
plt.tight_layout()
plt.show()
```
